```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import altair as alt
```

```python
# Bảng thống kê "Doanh thu" từng phim ở từng khu vực
dt_df = pd.read_excel('Downloads//DoanhThu.xlsx')
dt_df
```

```python
# Bảng thống kê Doanh thu ở Mỹ và Canada
dt1 = dt_df[['Phim','Doanh thu ở Mỹ và Canada ($)']].drop_duplicates()
dt1
```

```python
dt4_df = pd.read_excel('Downloads//dt4.xlsx')
dt4_df
```

```python
# Biểu đồ "Doanh thu ở Mỹ và Canada"
alt.Chart(dt1).mark_bar().encode(
 x = 'Phim:N',
 y = 'Doanh thu ở Mỹ và Canada ($):Q',
).properties(width = 550)
```

```python
# Top 5 phim có doanh thu cao nhất ở "Mỹ và Canada"
alt.Chart(dt4_df.head()).mark_bar().encode(
 y = 'Phim:N',
 x = 'Doanh thu ở Mỹ và Canada ($):Q',
).properties(width = 300)
```

```python
# Bảng thống kê doanh thu ở "Khu vực khác"
dt2 = dt_df[['Phim','Doanh thu ở khu vực khác ($)']].drop_duplicates()
dt2
```

```python
dt1_df = pd.read_excel('Downloads//DT1.xlsx')
dt1_df
```

```python
# Biểu đồ tổng quan doanh thu ở khu vực khác
alt.Chart(dt2).mark_bar().encode(
 x = 'Phim:N',
 y = 'Doanh thu ở khu vực khác ($):Q',
).properties(width = 500).interactive()
```

```python
#Top 5 phim có doanh thu ở "Khu vực khác" cao nhất
alt.Chart(dt1_df.head()).mark_bar().encode(
 y ='Phim:N',
 x ='Doanh thu ở khu vực khác ($):Q',
).properties(width = 300).interactive()
```
